```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
#BigQuery Extract-- 
alldata = pd.read_csv('ttime.csv')
zscore = np.abs(stats.zscore(alldata["tag_time"]))
# Shape of data before removing outliers
alldata.shape
#Removed outliers anything 3 or more standard deviations 
outliers_removed = alldata[(zscore < 3)]
outliers_removed.shape
#Average tag_load time per tag_name and sort and reset index
average_tag = outliers_removed.groupby('tag_name').agg(avg_time =('tag_time', 'mean'))
tag_time = average_tag.sort_values(['avg_time'], ascending=[False])
tag_time_reset = tag_time.reset_index()
#Set seaborn style
sns.set(style="whitegrid")
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12, 30))
# Plot
sns.set_color_codes("pastel")
sns.barplot(x="avg_time", y="tag_name", data=tag_time_reset,
            label="Avg.Time Seconds", color="b")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 8), ylabel="",
       xlabel="Avg.tag load speed excluding outliers 16/10-22/10")
sns.despine(left=True, bottom=True)
fig = ax.get_figure()
fig.savefig('e.png')






```

```python
allurls = pd.read_csv('urldata.csv')
#print(allurls.head(2))
zscore = np.abs(stats.zscore(allurls["tag_time"]))
#print(allurls.shape)
#Removed outliers anything 3 or more standard deviations 
outliers_removed_urls = allurls[(zscore < 3)]
outliers_removed_urls.shape

# Filter df for homepage
homepage = outliers_removed_urls[outliers_removed_urls['url']=='http://www.three.co.uk/']
homepageavg = homepage.groupby('tag_name').agg(avg_time =('tag_time', 'mean'))
hporder = homepageavg.sort_values(['avg_time'], ascending=[False])
hpreset = hporder.reset_index()

sns.set(style="whitegrid")
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))
# Plot
sns.set_color_codes("pastel")
sns.barplot(x="avg_time", y="tag_name", data=hpreset,
            label="Avg.Time Seconds", color="b")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 8), ylabel="",
       xlabel="Avg.tag load speed excluding outliers 16/10-22/10")
sns.despine(left=True, bottom=True)
fig = ax.get_figure()

# URLs to focus on #Focus on Clicktale
#clicktale = outliers_removed[(outliers_removed.tag_name == "clicktale")]
#sns.displot(x="avg_time", kind="ecdf", data = tag_time_reset)
#plt.xlim(0, 8)
```

```python
basket = outliers_removed_urls[outliers_removed_urls['url']=='https://store.three.co.uk/view/content/mbasket']
basketavg = basket.groupby('tag_name').agg(avg_time =('tag_time', 'mean'))
basketorder = basketavg.sort_values(['avg_time'], ascending=[False])
basketreset = basketorder.reset_index()

sns.set(style="whitegrid")
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))
# Plot
sns.set_color_codes("pastel")
sns.barplot(x="avg_time", y="tag_name", data=basketreset,
            label="Avg.Time Seconds", color="b")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 8), ylabel="",
       xlabel="Avg.tag load speed excluding outliers 16/10-22/10")
sns.despine(left=True, bottom=True)
fig = ax.get_figure()
```
